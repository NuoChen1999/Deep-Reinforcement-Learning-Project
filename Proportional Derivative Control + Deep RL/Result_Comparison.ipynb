```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
PD_file = '~/EE_positions/EE_positions_PD.csv'
PD_data = pd.read_csv(PD_file)

PD_RL_file = '~/EE_positions/EE_positions_PD_RL_20K.csv'
PD_RL_data = pd.read_csv(PD_RL_file)

target_pos = np.array([0.5, 0.0, 0.6])
steps = np.arange(len(PD_data))

plt.figure(figsize=(10, 6))

# X Position (red)
plt.plot(steps, PD_RL_data.iloc[:, 0], label="PD-RL x", color='red', linestyle='-')
plt.plot(steps, PD_data.iloc[:, 0], label="PD x", color='red', linestyle=':')
plt.plot(steps, np.full(len(steps), target_pos[0]), label="Target x", color='red', linestyle='--')

# Y Position (green)
plt.plot(steps, PD_RL_data.iloc[:, 1], label="PD-RL y", color='green', linestyle='-')
plt.plot(steps, PD_data.iloc[:, 1], label="PD y", color='green', linestyle=':')
plt.plot(steps, np.full(len(steps), target_pos[1]), label="Target y", color='green', linestyle='--')

# Z Position (blue)
plt.plot(steps, PD_RL_data.iloc[:, 2], label="PD-RL z", color='blue', linestyle='-')
plt.plot(steps, PD_data.iloc[:, 2], label="PD z", color='blue', linestyle=':')
plt.plot(steps, np.full(len(steps), target_pos[2]), label="Target z", color='blue', linestyle='--')

plt.legend(loc=9, bbox_to_anchor=(1.15, 0.75), ncol=1, frameon=False, fontsize='15')
plt.title('End-Effector Positions Comparison', fontsize='15')
plt.xlabel('Step', fontsize='15')
plt.ylabel('Position (m)', fontsize='15')
plt.grid(True)
plt.tight_layout()
plt.show()
```
